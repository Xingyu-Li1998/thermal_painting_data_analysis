```python
import numpy as np
import pandas as pd
```

```python
def read_csv(filename):
    df = pd.read_csv(filename)
    df = pd.DataFrame(df)
    df.columns = df.iloc[3]
    df = df.drop(3)
    df = df.iloc[3:]
    df = df.reset_index(drop=True)
    return df
filename = "thermal_painting_data/AFFDEX_Statistics_04.csv"
df_04 = read_csv(filename)
df_05 = read_csv("thermal_painting_data/AFFDEX_Statistics_05.csv")
df_combine = read_csv("thermal_painting_data/AFFDEX_Statistics_combine.csv")
df_raw_data_statistics = read_csv("thermal_painting_data/AFFDEX_Raw_Data_Statistics_annotation.csv")
```

```python
def read_csv_raw(filename):
    df = pd.read_csv(filename)
    df = pd.DataFrame(df)
    df.columns = df.iloc[1]
    df = df.drop(1)
    df = df.iloc[1:]
    df = df.reset_index(drop=True)
    return df
df_raw_data_statistics = read_csv_raw("thermal_painting_data/AFFDEX_Raw_Data_Statistics_annotation.csv")
```

```python
#compare the results from two cameras
df_compare = pd.concat([df_04,df_05],axis=0)
#store the different data
df_diff = pd.DataFrame()
def compare_camera(diff_df,df):
    for col in df.columns:
        if df[col].iloc[0] != df[col].iloc[1]:
            diff_df[col] = df[col]
    return diff_df

df_diff = compare_camera(df_diff,df_compare)
df_diff
#the angles of the cameras impact a lot
```
